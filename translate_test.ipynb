```python
from deep_translator import GoogleTranslator
import json
import os

# Path file hasil sebelumnya
input_json = "output/cleaned_text/epub_chunks.json"
output_json = "output/cleaned_text/epub_chunks_translated.json"

# Baca data asli
with open(input_json, "r", encoding="utf-8") as f:
    data = json.load(f)

translated_data = {}

for book_title, chunks in data.items():
    print(f"Translating: {book_title}")
    translated_chunks = []
    for chunk in chunks:
        try:
            translated = GoogleTranslator(source='auto', target='en').translate(chunk)
            translated_chunks.append(translated)
        except Exception as e:
            print("Gagal translate chunk:", e)
            translated_chunks.append("[TRANSLATION FAILED]")
    translated_data[book_title] = translated_chunks

# Simpan hasil
with open(output_json, "w", encoding="utf-8") as f:
    json.dump(translated_data, f, ensure_ascii=False, indent=2)

print("✅ Semua teks berhasil diterjemahkan dan disimpan ke:", output_json)
```
